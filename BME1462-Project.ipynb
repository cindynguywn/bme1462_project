1. Gaussian Filtering Image

```python
from skimage import io
from skimage import exposure, filters, io, morphology
animage = io.imread('test_image_1.tif')
```

```python
from skimage import filters
```

```python
import matplotlib.pyplot as plt
animage_sigma1 = filters.gaussian(animage, sigma=1, preserve_range=True)

animage_sigma10 = filters.gaussian(animage, sigma=10)
fig, axs = plt.subplots(1,2)
axs[0].imshow(animage, cmap='Greys_r')
axs[1].imshow(animage_sigma1, cmap='Greys_r')
plt.show()
```

```python
import numpy
# Gaussian smoothing to facilitate edge detection.
animage_smooth = filters.gaussian(animage, sigma=2, preserve_range=True)

# Contrast stretch.
animage_rescaled = exposure.rescale_intensity(animage_smooth, out_range=numpy.uint8)

# Local threshold.
amask = animage_rescaled >= filters.threshold_local(animage_rescaled, 33, method='gaussian')

# Display image and mask.
fig, axs = plt.subplots(1, 4)
axs[0].imshow(animage, cmap='Greys_r')
axs[1].imshow(animage_smooth, cmap='Greys_r')
axs[2].imshow(animage_rescaled, cmap='Greys_r')
axs[3].imshow(amask, cmap='binary_r')
plt.show()
```
